```python
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from tensorflow import keras
cancer = load_breast_cancer()
data = cancer['data']
target = cancer['target']
(train_data, test_data, train_label, test_label) = \
train_test_split(data,target,test_size=0.2,random_state=1)

#Dense = Fully Connected layer
model = keras.Sequential([
                          keras.layers.Dense(10, activation='relu', input_shape=(30,)),
                          keras.layers.Dense(1, activation='sigmoid')
])
#loss (손실함수) : 어떤 방식으로 에러값을 구할래
#optimizer : 에러가 생겼어. 어떤 방식으로 weight을 바꿔줄래
#metrics = 어떤걸 더 보고싶니
model.compile(loss='BinaryCrossentropy', optimizer='adam',metrics=['accuracy'])

#batch_size : 몇개씩 훈련시켜서 weight값을 바꿀건지
#epochs : 몇번 훈련시킬거니 (너무 많으면(overfittnig) 또 안됨 )
# ex) 10000개의 데이터를 16개씩 쪼개서 10000/16 번 훈련시킨게 1 epoch
model.fit(train_data,train_label,batch_size=16,epochs=20)
loss, acc = model.evaluate(test_data, test_label)
print(acc)
```

```python
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from tensorflow import keras
cancer = load_breast_cancer()
data = cancer['data']
target = cancer['target']
(train_data, test_data, train_label, test_label) = \
train_test_split(data,target,test_size=0.2,random_state=1)

#Dense = Fully Connected layer
model = keras.Sequential([
                          keras.layers.Dense(30, activation='relu', input_shape=(30,)),
                          keras.layers.Dense(30, activation='relu',),
                          keras.layers.Dense(30, activation='relu',),
                          keras.layers.Dense(30, activation='relu',),
                          keras.layers.Dense(1, activation='sigmoid')
])
#loss (손실함수) : 어떤 방식으로 에러값을 구할래
#optimizer : 에러가 생겼어. 어떤 방식으로 weight을 바꿔줄래
#metrics = 어떤걸 더 보고싶니
model.compile(loss='BinaryCrossentropy', optimizer='adam',metrics=['accuracy'])

#batch_size : 몇개씩 훈련시켜서 weight값을 바꿀건지
#epochs : 몇번 훈련시킬거니 (너무 많으면(overfittnig) 또 안됨 )
# ex) 10000개의 데이터를 16개씩 쪼개서 10000/16 번 훈련시킨게 1 epoch
model.fit(train_data,train_label,batch_size=16,epochs=30)
loss, acc = model.evaluate(test_data, test_label)
print(acc)
```
